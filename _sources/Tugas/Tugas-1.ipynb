Memasukan Rumus Menghitung Jarak Data

```python
DECREMENT_BY_ONE = 1
INCREMENT_BY_ONE = 1

PROBALITY_TABLE_VALUE = {
    "q" : (1,1),
    "r" : (1,0),
    "s" : (0,1),
    "t" : (0,0),
}
```

```python
def get_series(data_sbmptn, idx, series):
    return data_sbmptn.loc[(idx), series]
```

```python
def dissimilarity_dataset(data_sbmptn, series_index = [], series = []):
    first_series = get_series(data_sbmptn, series_index[0], series)
    second_series = get_series(data_sbmptn, series_index[1], series)
    dataset = pd.concat([first_series,second_series],axis=1)
    return dataset.T
```

Hitung hasil jarak jangkauan tipe data

```python
def hasil_hitung_jarak(data_sbmptn, start_index = 0, last_index = 1):

    PROBALITY_VALUE = {"q" : 0, "r" : 0, "s" : 0, "t" : 0}

    column_range = data_sbmptn.shape[1]

    for column in range(column_range):
        for value in PROBALITY_TABLE_VALUE:
            item = list((tuple(data_sbmptn.loc[(start_index):(last_index), data_sbmptn.columns[column]]) == PROBALITY_TABLE_VALUE[value], value))
            if item[0] == True:
                if item[1] == "q":
                    PROBALITY_VALUE["q"] += 1
                if item[1] == "r":
                    PROBALITY_VALUE["r"] += 1
                if item[1] == "s":
                    PROBALITY_VALUE["s"] += 1
                if item[1] == "t":
                    PROBALITY_VALUE["t"] += 1

    return PROBALITY_VALUE
```

```python
#(1,2)
dimensi_1_2 = dissimilarity_dataset(data_sbmptn, [1,2], ["jenis_kelamin", "kategori_jurusan"])
```

```python
probality_d_1_2 = hasil_hitung_jarak(dimensi_1_2, 1, 2)
```

```python
#(1,3)
dimensi_1_3 = dissimilarity_dataset(data_sbmptn, [1,3], ["jenis_kelamin", "kategori_jurusan"])
```

```python
probality_d_1_3 = hasil_hitung_jarak(dimensi_1_3, 1, 3)
```

```python
#(1,4)
dimensi_1_4 = dissimilarity_dataset(data_sbmptn, [1,4], ["jenis_kelamin", "kategori_jurusan"])
```

```python
probality_d_1_4 = hasil_hitung_jarak(dimensi_1_4, 1, 4)
```

Perhititungan Rumus dengan metode Dissimilarity Binary Symetric

```python
def dissimilarity_jarak_simetris_binary(probality_value):
  
    return (probality_value["r"] + probality_value["s"]) / (probality_value["q"] + probality_value["r"] + probality_value["s"] + probality_value["t"])
```

```python
dimensi_1_2 = dissimilarity_jarak_simetris_binary(probality_d_1_2)
dimensi_1_3 = dissimilarity_jarak_simetris_binary(probality_d_1_3)
dimensi_1_4 = dissimilarity_jarak_simetris_binary(probality_d_1_4)
```
