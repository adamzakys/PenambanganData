## Naive Bayes

### Pertama Import beberapa modul python yang diperlukan





```python
import pandas as pd
import numpy as np
```

```python
datasetscr = "https://raw.githubusercontent.com/kb22/Heart-Disease-Prediction/master/dataset.csv"
```

```python
df = pd.read_csv(datasetscr)
df.head()
```

cek jumlah baris dan kolom yang ada di dataset

```python
df.shape
```

### Menghitung Data

Pada Langkah ini pisahkan kolom "target" dari data frame dan ambil kolom "target" sebagai target kolom untuk mengkategorikan

Selanjutnya menentukan variabel independen dan variabel dependen dari data yang akan dianalisis

```python
x = df.drop(['target'], axis = 1)
x.head()
```

Pada Kolom result of treatment di drop atau di hapus dari data frame karena akan menjadi variabel dependen

```python
y = df['target']
y.head()
```

Selanjutnta melakukan analisis menggunakan klasifikasi naive bayes. untuk langkah dengan membagi dataset menjadi training set dan test set

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 123)
```

Di program atas saya membagi dataset menjadi 80% training set dan 20% test set

Lalu memisahkan untuk melakukan prediksi pada training set dan test set

```python
# Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
# Mengaktifkan/memanggil/membuat fungsi klasifikasi Naive bayes
modelnb = GaussianNB()
# Memasukkan data training pada fungsi klasifikasi naive bayes
nbtrain = modelnb.fit(x_train, y_train)
nbtrain.class_count_
```

Menentukan hasil prediksi dari x_test

```python
y_pred = nbtrain.predict(x_test)
y_pred
```

Menentukan probbabilitas prediksi

```python
nbtrain.predict_proba(x_test)
```

Setelah memperoleh hasil nilai prediksi y_pred, tahapan lanjutnya adalah melakukan Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

Langkah terakhir yaitu melakukan perhitungan hasil nilai akurasi

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
