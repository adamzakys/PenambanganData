## Deskripsi Aplikasi :

*   Aplikasi ini bertujuan untuk memprediksi kemungkinan seorang pendonor akan mendonorkan darahnya pada saat kendaraan datang ke kampus berikutnya. Dengan menggunakan data seperti recency (bulan sejak donasi terakhir), frequency (jumlah donasi), monetary (total darah yang didonorkan dalam cc), dan time (bulan sejak donasi pertama), aplikasi ini mengimplementasikan beberapa metode pemodelan, termasuk Naive Bayes, KNN, Decision Tree, dan MLP.
*   Langkah awal melibatkan pemahaman data, diikuti oleh proses preprocessing untuk mengatasi kesalahan atau ketidaklengkapan data. Salah satu tahap dalam preprocessing adalah normalisasi data, yang merupakan elemen dasar dalam data mining untuk memastikan konsistensi dalam dataset.
*   Dalam tahap modelling, aplikasi menawarkan berbagai pilihan model seperti Naive Bayes, KNN, Decision Tree, dan MLP. Pengguna dapat memilih model yang paling sesuai dengan kebutuhan analisis mereka.
*   Pada tahap implementasi, pengguna diminta untuk memasukkan nilai recency, frequency, monetary, dan time. Berdasarkan input tersebut, aplikasi memberikan prediksi kemungkinan pendonor untuk mendonorkan darah pada kunjungan kendaraan berikutnya.
*   Aplikasi ini merupakan alat yang berguna untuk membantu dalam perencanaan kegiatan donor darah di kampus dengan memberikan prediksi potensial dari calon pendonor. Seluruh proses, mulai dari pemahaman data, preprocessing, modelling, hingga implementasi, dapat dilakukan melalui antarmuka yang disajikan dengan bantuan Streamlit.
